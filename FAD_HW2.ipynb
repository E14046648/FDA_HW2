```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO   
from sklearn.tree import export_graphviz
import pydotplus
import random
```

```python
df = pd.DataFrame(columns=["IQ", "study_hour", "class_attent", "homework", "gender","pass"])
df['IQ']=[None]*1000
df.iloc[1,0]=10
df
```

```python
i = 0 
while i<1000:
    df.iloc[i,0]=random.randint(80,180)
    df.iloc[i,1]=random.randint(0,8)
    df.iloc[i,2]=random.randint(0,18)
    df.iloc[i,3]=random.randint(0,100)
    df.iloc[i,4]=random.randint(0,1)
    if df.iloc[i,4]==0:  ##gender is girl
        df.iloc[i,5]=1  #pas
    else:
        if df.iloc[i,0]>150:
            if df.iloc[i,2]>3:
                df.iloc[i,5]=1
            else:
                df.iloc[i,5]=0
        elif df.iloc[i,0]<100:
            df.iloc[i,5]=0
        else:
            if  df.iloc[i,1]>4:#study_hour more than 4h/week
                df.iloc[i,5]=1
            else:
                if df.iloc[i,3]>60:
                    df.iloc[i,5]=1
                else:
                    df.iloc[i,5]=0
                
                
    i=i+1
```

```python
df
df.to_excel('E:/Anaconda/file/tree1.xls')
```

```python
df_train = df[:700]
df_train.head()
```

```python
df_test = df[700:]
df_test.head()
```

```python
y = df_train['pass'].values
y
```

```python
df_train = df_train.drop('pass', 1)
```

```python
dtree=DecisionTreeClassifier(max_depth=4)
dtree.fit(df_train,y)

dot_data = StringIO()
export_graphviz(dtree, 
                out_file=dot_data,  
                filled=True, 
                feature_names=list(df_train),
                class_names=['fail','pass'],
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_pdf("E:/Anaconda/file/tree1.pdf")
```

```python
y_test = df_test['pass'].values
X_test = df_test.drop('pass', 1)

y_predict = dtree.predict(X_test)

y_predict
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict)
```
